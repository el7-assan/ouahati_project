```python
months = ["يناير", "فبراير", "مارس", "أبريل", "ماي", "يونيو", "يوليوز", "غشت", "سبتمبر", "أكتوبر", "نوفمبر", "ديسمبر"]

monthly_crops = {m: [] for m in months}

for crop, info in crops_data.items():
    season = info.get("season", "").lower()
    
    # شتوي
    if "شتوي" in season:
        for m in ["أكتوبر", "نوفمبر", "ديسمبر", "يناير", "فبراير", "مارس"]:
            if crop not in monthly_crops[m]:
                monthly_crops[m].append(crop)
    
    # ربيعي أو صيفي
    if "ربيعي" in season or "صيفي" in season:
        for m in ["مارس", "أبريل", "ماي", "يونيو", "يوليوز", "غشت", "سبتمبر"]:
            if crop not in monthly_crops[m]:
                monthly_crops[m].append(crop)
    
    # سنوي
    if "سنوي" in season:
        tree_crops = ["تمر", "نخيل", "زيتون", "تين", "رمان", "لوز", "حناء", "ورد", "ليمون", "صبار"]
        if any(tc in crop for tc in tree_crops):
            for m in ["مارس", "أبريل", "ماي"]:
                if crop not in monthly_crops[m]:
                    monthly_crops[m].append(crop)
        else:
            # طوال السنة
            for m in months:
                if crop not in monthly_crops[m]:
                    monthly_crops[m].append(crop)

# ترتيب القوائم + حفظ
for m in monthly_crops:
    monthly_crops[m] = sorted(set(monthly_crops[m]))  # بدون تكرار

with open('months.pkl','wb')as f:
    pickle.dump(monthly_crops,f)
```
